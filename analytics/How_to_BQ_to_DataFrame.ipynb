Construct BigQuery Query to retrieve data

```python
query = """
select *
from job_search.job_details
where search_datetime between '2021-01-01' and '2021-01-03'
and lower(title) like '%data%scientist%'
"""
```

Instantiate and retrieve query results to a dataframe

```python
bq_client = bigquery.Client()
df = bq_client.query(query).result().to_dataframe()
```
